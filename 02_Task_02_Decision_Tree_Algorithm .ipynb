# Prediction using Decision Tree  Algorithm 
 
* Create the Decision Tree classifier and visualize it graphically. 

* The purpose is if we feed any new data to this classifier, it would be able to  predict the right class accordingly.  
* ***Dataset*** : https://bit.ly/3kXTdox 

## Importing Modules

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, export_graphviz
import pandas as pd
import pydotplus
import numpy as np
from IPython.display import Image
import os
os.environ["PATH"] += os.pathsep + r'C:\Program Files\Graphviz\bin'
```

##  Loading Data

```python
iris = load_iris()
```

## Creating a dataframe

```python
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df['target'] = iris.target
```

```python
# Let's see target names
targets = iris.target_names
print(targets)
```

## Prepare training data for building the model

```python
X_train = df.drop(['target'], axis=1)
y_train = df['target']
```

```python
# Instantiate the model
cls = DecisionTreeClassifier()
```

## Train/Fit the model 

```python
cls.fit(X_train, y_train)
```

## Make prediction using the model

```python
X_pred = [5.1, 3.2, 1.5, 0.5]
y_pred = cls.predict([X_pred])

print("Prediction is: {}".format(targets[y_pred]))
```

## Create DOT data

```python
dot_data = export_graphviz(cls, out_file=None, 
                                feature_names=iris.feature_names,  
                                class_names=targets,filled=True, rounded=True,
                                special_characters=True)
```

```python
# Draw graph
graph = pydotplus.graph_from_dot_data(dot_data)
graph
```

## Show graph

```python
Image(graph.create_png())
```
